```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fn='nino12_dropnan.txt'
df=pd.read_csv(fn)        
df_describe = df.describe()
maxValue = df_describe.at['max','Nino12']
minValue = df_describe.at['min','Nino12']
category = [minValue, -0.5, 0,0.5,maxValue]
labels =['LaNinaTemp', 'Cold', 'Warm', 'NinoTemp']
x = pd.cut(df['Nino12'],category,right=False,labels=labels).rename('Label')
df4 = pd.concat([df,x],axis=1)
df4.to_csv('nino12_dropnan_result.csv',index=False)
plt.figure()
plt.pie([i[1] for i in list(x.value_counts().items())],
labels=labels)
plt.show()

print('任务三完成')
```

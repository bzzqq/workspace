```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('4.pollution_us_5city_2006_2010_NO2.csv')
df_2007 = df[df['Date Local'].str.startswith('2007')].dropna()
df_2007.to_csv('pollution_us_5city_2007_NO2.txt', sep=' ', index=False)
df_txt = pd.read_csv('pollution_us_5city_2007_NO2.txt', sep=' ')
df_sorted = df_txt.sort_values(by='County Code', ascending=False)
df_sorted.to_csv('pollution_us_5city_2007_NO2_descending.csv', index=False)
new_df = pd.read_csv('pollution_us_5city_2007_NO2_descending.csv')
queens_df = new_df[new_df['County'] == "Queens"]
category = [0, 25, 50, 75, 100, 125, 150]
labels = ['Good', 'Moderate', 'SubUnhealthy', 'Unhealthy', 'VeryUnhealthy', 'Hazardous']
queens_df['NO2 AQI离散化'] = pd.cut(queens_df['NO2 AQI'], bins=category, labels=labels)
counts = queens_df['NO2 AQI离散化'].value_counts()
counts.plot(kind='bar'); plt.savefig('NO2_AQI_bar.png', dpi=300); plt.close()
counts.plot(kind='pie', autopct='%1.1f%%'); plt.savefig('NO2_AQI_pie.png', dpi=300); plt.close()
```

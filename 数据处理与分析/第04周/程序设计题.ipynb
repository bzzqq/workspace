```python
# 寻找货架 I
purchases = ["西瓜", "苹果", "洗洁精", "芹菜", "牙膏", "香蕉"]
shelf1 = ["西瓜", "苹果", "梨", "香蕉", "草莓"]
shelf2 = ["西红柿", "黄瓜", "土豆", "芹菜", "韭菜"]
shelf3 = ["洗衣液", "洗洁精", "肥皂", "洗发水", "牙膏"]
for item in purchases:
    if item in shelf1:
        print(f"您购买的{item}在1号货架。")
    elif item in shelf2:
        print(f"您购买的{item}在2号货架。")
    else:
        print(f"您购买的{item}在3号货架。")
```

```python
# 寻找货架 II
purchases = ["西瓜", "苹果", "洗洁精", "芹菜", "牙膏"]
shelves = {
    "1号货架": ["西瓜", "苹果", "梨", "香蕉", "草莓"],
    "2号货架": ["西红柿", "黄瓜", "土豆", "芹菜", "韭菜"],
    "3号货架": ["洗衣液", "洗洁精", "肥皂", "洗发水", "牙膏"],
}
for item in purchases:
    for shelf, stock in shelves.items():
        if item in stock:
            print(f"您购买的{item}在{shelf}。")
```

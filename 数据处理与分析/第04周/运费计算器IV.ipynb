```python
import math

def calculate_cost(destination, weight):
    # 定义不同层级的目的地
    tier1 = ["江苏", "浙江", "上海"]
    tier2 = [
        "安徽", "广东", "福建", "山东", "江西", "北京", "天津", "河南", "河北", "湖南", "湖北", 
        "海南", "陕西", "辽宁", "云南", "四川", "重庆", "贵州", "宁夏", "吉林", "甘肃", "青海", 
        "黑龙江", "山西", "广西"
    ]
    tier3 = ["内蒙古", "新疆"]
    
    # 定义不同层级的运费规则
    tier_costs = {
        'tier1': 8,
        'tier2': {'base': 15, 'per_kg': 12},
        'tier3': {'base': 20, 'per_kg': 18}
    }
    
    # 根据目的地确定层级
    if destination in tier1:
        cost = tier_costs['tier1']
    elif destination in tier2:
        cost_details = tier_costs['tier2']
        # 向上取整计算重量
        weight_ceil = math.ceil(weight)
        cost = cost_details['base'] + (weight_ceil - 1) * cost_details['per_kg'] if weight_ceil > 1 else cost_details['base']
    else:
        cost_details = tier_costs['tier3']
        # 向上取整计算重量
        weight_ceil = math.ceil(weight)
        cost = cost_details['base'] + (weight_ceil - 1) * cost_details['per_kg'] if weight_ceil > 1 else cost_details['base']
    
    return cost

# 运费数据
delivery = {"上海": 8, "广东": 0.8, "江西": 1.8, "河北": 2, "内蒙古": 2, "重庆": 4}

# 计算并打印每个目的地的运费
for destination, weight in delivery.items():
    cost = calculate_cost(destination, weight)
    print("=============")
    print(f"目的地：{destination}")
    print(f"重量：{weight:.1f} kg")
    print(f"价格：{cost:.1f} 元")
print("=============")
```

```python
print("=== 聊天机器人 ===")
user_input = input("你说：")

# 多条件匹配
if "你好" in user_input or "hi" in user_input.lower() or "hello" in user_input.lower():
    print("机器人：你好！很高兴见到你")

elif "天气" in user_input:
    if "下雨" in user_input:
        print("机器人：下雨天记得带伞哦")
    elif "晴天" in user_input:
        print("机器人：晴天适合出去玩")
    else:
        print("机器人：今天的天气还不错")

elif "名字" in user_input or "谁" in user_input:
    if "你" in user_input and "名字" in user_input:
        print("机器人：我叫小智")
    else:
        print("机器人：我是你的聊天助手")

elif "吃饭" in user_input or "饿了" in user_input:
    if "中午" in user_input or "12" in user_input:
        print("机器人：该吃午饭了，记得按时吃饭")
    elif "晚上" in user_input or "18" in user_input:
        print("机器人：晚饭时间到，别饿着")
    else:
        print("机器人：饿了就去吃点东西吧")

elif "喜欢" in user_input and "什么" in user_input:
    if "音乐" in user_input:
        print("机器人：我喜欢轻音乐")
    elif "电影" in user_input:
        print("机器人：我喜欢科幻电影")
    else:
        print("机器人：我喜欢和大家聊天")

elif "?" in user_input or "？" in user_input:
    print("机器人：这个问题让我想想...")
    print("机器人：抱歉，我还不太明白")

else:
    print("机器人：嗯嗯，然后呢？")
```

```python
weather = "晴"
temperature = 25

# 如果明天晴天且温度小于30度，那么去户外运动
if weather == "晴" and temperature < 30:
    print("完美天气！适合户外运动")

# 如果明天不是晴天或气温低于10度，就窝在被窝里
if weather != "晴" or temperature < 10:
    print("适合窝在被窝里")
```

```python
# 敏感词过滤
sensitive_words = "赌博"
comment = input("请输入评论内容：")

# 检查是否包含敏感词
if sensitive_words in comment:
    print("评论包含敏感词！")
    print("评论发布失败，请修改后重试")
else:
    print("评论发布成功")
```

```python
# 敏感词过滤
sensitive_words = "赌博"
comment = input("请输入评论内容：")

# 检查是否包含敏感词
if sensitive_words not in comment:
    print("评论发布成功")
else:
    print("评论包含敏感词！")
    print("评论发布失败，请修改后重试")
```

```python
year = int(input("请输入年份："))

# 判断闰年
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"{year}年2月有29天")
else:
    print(f"{year}年2月有28天")
```

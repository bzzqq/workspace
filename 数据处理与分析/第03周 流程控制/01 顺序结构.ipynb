```python
num = int(input())
if num <= 1:
    print('请输入一个大于1的自然数')
elif num <= 3:
    print(f'{num}是素数')
else:
    for i in range(2, num):
        if num % i == 0:
            print(f'{num}={i}*{num // i},{num}不是素数')
            break
    else:
        print(f'{num}是素数')
```

```python
i = 1
sign = 1
pi = 0
while 1 / i >= 0.000001:
    pi = pi + 1 / i * sign
    sign = sign * (-1)
    i = i + 2
print(4 * pi)
```

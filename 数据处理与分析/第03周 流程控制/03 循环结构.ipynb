```python
# 猜数字游戏
lucky_number = (((5 + 7) * 2) - (8 * 3 // 2)) // (4 - 2)
guess_number = int(input())
while guess_number != lucky_number:
    print("很遗憾，没有猜对！")
    guess_number = int(input())
print("恭喜你，猜中了！")
```

```python
fruits = ['apple', 'banana', 'pear']  # 定义一个列表
user_input = input('请输入一个水果名称')  # 提示用户输入
while user_input not in fruits:  # 检查用户输入是否在列表中
    print('输入无效，请重新输入！')
    user_input = input('请输入一个水果名称')
print(f'你输入的水果是：{user_input}')
```

```python
for cock in range(1, 20):                   # 公鸡数量不超过20
    for hen in range(1, 33):                # 母鸡数量不超过33
        chicken = 100 - cock - hen        #小鸡数量可计算得到
        if chicken % 3 == 0 and 5 * cock + 3 * hen + chicken // 3 == 100:
            print(cock, hen, chicken)      # 遇到满足条件的数字组合就输出
```

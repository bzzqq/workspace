```python
# continue和break的区别
for num in range(1, 6):
    if num % 2 == 0:
        break  # 遇到偶数终止
    print(num)
```

```python
# continue和break的区别
for num in range(1, 6):
    if num % 2 == 0:
        continue  # 跳过偶数
    print(num)
```

```python
scores = [60, 70, 110, 80, 90, 120]
for score in scores:
    if score > 100:
        continue
    print(score)
```

```python
n = int(input())  # 输入正整数n,且n > 1
for i in range(2, n):  # i 依次取值2,3...n-1
    if n % i == 0:  # 能被2,3...n-1整除的数不是素数
        print(f'{n}不是素数')
        break  # 提前结束循环，跳过else子句
else:  # 循环中未遇到break时执行
    print(f'{n}是素数')
```

```python
friends = ['金吒', '木吒', '哪吒', '敖甲', '敖乙', '敖丙']
invited_friends = ['哪吒', '敖丙']
for friend in friends:
    if friend in invited_friends:
        continue
    print(f'邀请{friend}')
```

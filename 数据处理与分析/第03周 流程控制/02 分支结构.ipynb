# 多个if和if...elif...的区别

```python
# 运输方式选择
# 距离 < 200 货车；200 <= 距离 < 800，铁路；
# 800 <= 距离 < 1500，航空；距离 > 1500 海运
distance = int(input())
if distance < 200:
    transport = '货车'
```

```python
distance = int(input())
if distance < 200:
    transport = '货车'
if 200 <= distance < 800:
    transport = '铁路'
if 800 <= distance < 1500:
    transport = '航空'
if distance > 1500:
    transport = '海运'
print(transport)
```

```python
distance = int(input())
if distance < 200:
    transport = '货车'
elif 200 <= distance < 800:
    transport = '铁路'
elif distance < 1500:
    transport = '航空'
else:
    transport = '海运'
print(transport)
```

```python
# 多条件嵌套判断
distance = int(input())
is_dangerous = False
if distance > 800:
    transport = '航空'
    if is_dangerous:
        transport = '陆运'
else:
    transport = '陆运'
print(transport)
```

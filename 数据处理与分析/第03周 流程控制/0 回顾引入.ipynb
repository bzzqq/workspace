```python
import math

a = 1
b = 2
c = 1
# 计算判别式
delta = b**2 - 4 * a * c
# 根据判别式判断根的情况
if delta >= 0:
    x1 = (-b + math.sqrt(delta)) / (2 * a)
    x2 = (-b - math.sqrt(delta)) / (2 * a)
    print(x1, x2)
else:
    print("方程没有实数解")
```

```python
import math

# 计算绳子与地球的空隙
r = 6371000  # 地球半径（米）
l = 2 * math.pi * r  # 赤道周长
gap = (l + 1) / (2 * math.pi) - r  # 绳子与地球的空隙（米）
print(f"空隙高度: {gap:.3f} 米 ({gap*100:.1f} 厘米)\n")

# 定义10种动物的高度（单位：米）
animals = {
    "老鼠": 0.10,  # 老鼠高度约10厘米
    "小猫": 0.25,    # 猫的肩高约25厘米
    "小狗": 0.40,    # 中型狗的肩高约40厘米
    "兔子": 0.20,  # 兔子的肩高约20厘米
    "大象": 2.50,  # 大象肩高约2.5米
    "企鹅": 0.70,  # 帝企鹅高度约70厘米
    "猴子": 0.40,  # 猴子的肩高约40厘米
    "熊猫": 0.70, # 大熊猫肩高约70厘米
    "仓鼠": 0.05,  # 仓鼠高约5厘米
    "长颈鹿": 3.00 # 长颈鹿肩高约3米（不算脖子和头）
}

# 判断每种动物能否通过
print("动物能否通过空隙？")
print("-" * 30)

for animal, height in animals.items():
    if gap >= height:
        print(f"{animal}可以通过空隙✅")
    else:
        print(f"{animal}无法通过空隙❌")
```

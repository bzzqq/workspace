```python
# 返回一个随机物品
import random as r

item_pool = ["铜币", "治疗药水", "铁剑", "皮革护甲", "银币", "魔力药水", "钢制长剑", "链甲"]
item = r.choice(item_pool)
print(item)
```

```python
# 返回多个随机物品
import random as r

item_pool = ["铜币", "治疗药水", "铁剑", "皮革护甲", "银币", "魔力药水", "钢制长剑", "链甲"]
item = r.choices(item_pool, k=3)
print(item)
```

```python
import random

# 定义道具池（可根据需求扩展）
item_pool = {"普通": ["铜币", "治疗药水", "铁剑", "皮革护甲"], "稀有": ["银币", "魔力药水", "钢制长剑", "链甲"], "史诗": ["金币", "复活卷轴", "秘银匕首", "板甲"], "传说": ["钻石", "龙之心脏", "圣剑Excalibur", "神谕长袍"]}

# 设置各级别掉落概率（总和建议为1）
probabilities = {
    "普通": 0.65,  # 65%
    "稀有": 0.25,  # 25%
    "史诗": 0.08,  # 8%
    "传说": 0.02  # 2%
}

rarity = random.choices(list(probabilities.keys()), weights=list(probabilities.values()), k=1)
print(rarity)
print(item_pool[rarity[0]])

# 从对应稀有度池中随机选择道具
item = random.choice(list(item_pool[rarity[0]]))
print(item)

print(f"获得{rarity[0]}级道具：{item}")
```

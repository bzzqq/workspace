```python
# 谁得了MVP？
# 生成包含10个同学的成绩字典，中文姓名为键，成绩为值。
students = {"小明": 90, "小红": 85, "小刚": 92, "小李": 88, "小赵": 95, "小钱": 87, "小孙": 91, "小周": 83, "小吴": 99, "小郑": 86}


# 返回所有90分以上学生的字典
def get_mvp(students):
    mvp = {}
    for name, score in students.items():
        if score >= 90:
            mvp[name] = score
    return mvp


print(get_mvp(students))
```

```python
# 智能库存预警系统
def inventory_alert(current_stock, daily_sales):
    safety_stock = daily_sales * 3
    if current_stock < daily_sales * 1.5:
        print("库存即将耗尽，建议紧急补货")
    elif current_stock < safety_stock:
        print("低于安全库存，建议补货")
    else:
        print("库存充足")


inventory_alert(100, 80)
```

```python
# 自动化分拣系统
packages = {
    '茶杯': {
        'weight': 0.5,
        'value': 10,
        'material': 'glass',
        'freezed': False
    },
    '项链': {
        'weight': 0.1,
        'value': 2500,
        'material': 'gold',
        'freezed': False
    },
    '书': {
        'weight': 0.2,
        'value': 100,
        'material': 'paper',
        'freezed': False
    },
    '车厘子': {
        'weight': 2.0,
        'value': 100,
        'material': 'fruit',
        'freezed': True
    },
    '冰棒': {
        'weight': 0.1,
        'value': 10,
        'material': 'ice',
        'freezed': True
    },
    '镜子': {
        'weight': 0.2,
        'value': 50,
        'material': 'glass',
        'freezed': False
    },
    '桌子': {
        'weight': 20,
        'value': 800,
        'material': 'wood',
        'freezed': False
    },
}


def package_sorting(properties):
    if properties['value'] > 2000:
        return '贵重区'
    elif properties['weight'] > 10:
        return '大件区'
    elif properties['material'] == 'glass':
        return '易碎区'
    elif properties['freezed']:
        return '冷链区'
    else:
        return '普通区'


for package, property in packages.items():
    print(package, package_sorting(property))
```

```python
# 订单金额统计
orders = {"id1": 158.5, "id2": 299.0, "id3": 45.8, "id4": 100.0, "id5": 300.0, "id6": 20.0}


def calculate_order_amount(orders):
    """
    返回最大金额订单
    返回最低金额订单
    返回订单总金额
    """
    max_amount = max(orders.values())
    min_amount = min(orders.values())
    total_amount = sum(orders.values())
    print("最大金额订单：", max_amount)
    print("最低金额订单：", min_amount)
    print("订单总金额：", total_amount)


calculate_order_amount(orders)
```

```python
# 基础运费10元，续重每1kg运费3元
weights = [1, 2, 3, 4, 5]
for weight in weights:
    cost = 10 + (weight - 1) * 3
```

```python
# 基础运费10元，续重每1kg运费3元，不足1kg按1kg计算
weights = [0.5, 1, 2, 3, 4]
for weight in weights:
    if weight <= 1:
        cost = 10
    else:
        cost = 10 + (weight - 1) * 3
    print('=================')
    print(f'包裹重量：{weight:.2f} kg')
    print(f'运费：{cost:.2f} 元')
print('=================')
```

```python
# 基础运费10元，续重：超过1kg小于等于5kg，每kg运费3元
# 超过5kg小于等于10kg，每kg运费2元，大于10 kg，每kg运费1.5元
# 不足1kg按1kg计算
weights = [0.5, 2, 6, 8, 11, 15]
for weight in weights:
    if weight <= 1:
        cost = 10
    elif 1 < weight <= 5:
        cost = 10 + (weight - 1) * 3
    elif 5 < weight <= 10:
        cost = 22 + (weight - 5) * 2
    else:
        cost = 32 + (weight - 5) * 1.5
    print(f"重量：{weight:.2f} kg，总运费：{cost:.2f}元")
```

```python
weights = [0.5, 1, 2, 3, 4]
for weight in weights:
    if weight <= 1:
        cost = 10
    else:
        cost = 10 + (weight - 1) * 3
    print(weight, cost, sep= ' ==> ')
```

```python
weights = [0.5, 1.5, 2, 3.5, 4]
for weight in weights:
    if weight <= 1:
        cost = 10
    elif weight % 1 != 0:
        cost = 10 + (weight // 1) * 3
    else:
        cost = 10 + (weight - 1) * 3
    print(weight, cost, sep=' ==> ')
```

```python
import numpy as np
weights = [0.5, 1.5, 2.5, 3, 4]
for weight in weights:
    if weight <= 1:
        cost = 10
    else:
        cost = 10 + np.ceil(weight) * 3
    print(weight, cost, sep=' ==> ')
```

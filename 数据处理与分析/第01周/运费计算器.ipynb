```python
weight = int(input())
distance = int(input())
price = 2 * weight * distance
print("========= 运费明细 =========")
print(f"货物重量：{weight} kg")
print(f"运输距离：{distance} 公里")
print(f"总费用：{price} 元")
print("============================")
print("货物重量：{} kg，运输距离：{} 公里".format(weight, distance))
```

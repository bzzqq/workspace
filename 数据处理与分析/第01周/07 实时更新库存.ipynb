```python
current_stock = 500  
# 入库操作  
current_stock = current_stock + 200
print("当前库存:", current_stock)  # 输出：700  
```

```python
current_stock = 500  
# 入库操作  
current_stock += 200    # 等价于 current_stock = current_stock + 200
print("当前库存:", current_stock)  # 输出：700  
```

```python
current_stock = 500  
# 入库操作  
current_stock += 200    # 等价于 current_stock = current_stock + 200  
# 出库操作  
current_stock -= 150  
print("当前库存:", current_stock)  # 输出：550  
```

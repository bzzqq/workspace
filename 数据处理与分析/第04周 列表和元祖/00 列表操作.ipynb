```python
# 列表可以保存任意数量、任意类型的数据
# 列表中的元素用逗号分割
# 建议给列表指定一个复数的名称
list1 = [1, 2, 3, 4, 5]
list2 = ['张', '王', '李', '赵', '钱']
list3 = [1, 2, '张', '王']
```

```python
# 访问列表元素
# 索引从0开始
# 负数索引从-1开始
names = ['张', '王', '李', '赵', '钱']
print(names[-1])
```

```python
# 修改、添加、删除列表元素
# 添加 append()、insert()
# 删除 del pop() remove()
names = ['张', '赵', '赵', '王', '李', '赵', '钱']
names.pop()
print(names)
```

```python
log = ["入库", "出库", "盘点"]
log.insert(0, "紧急出库") 
print(log)
print(log[1])
```

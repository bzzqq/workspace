```python
# 用图形理解索引
fruits = ["苹果", "香蕉", "橙子", "葡萄"]

print("正向索引：")
print("位置:   0      1      2      3")
print(f"内容: {fruits[0]}    {fruits[1]}    {fruits[2]}    {fruits[3]}")
print()
print("反向索引：")
print("位置:   -4     -3     -2     -1")
print(f"内容: {fruits[-4]}    {fruits[-3]}    {fruits[-2]}    {fruits[-1]}")
```

```python
names = ['иөө', 'й’ұ', 'еӯҷ', 'жқҺ', 'е‘Ё', 'еҗҙ', 'йғ‘', 'зҺӢ', 'еҶҜ', 'йҷҲ']
# иҺ·еҸ–з¬¬1гҖҒ3гҖҒ5гҖҒ7гҖҒ9пјҲжүҖжңүеҘҮж•°пјүдёӘе…ғзҙ 
print(names[::2])
# иҺ·еҸ–з¬¬2гҖҒ4гҖҒ6гҖҒ8гҖҒ10пјҲжүҖжңүеҒ¶ж•°пјүдёӘе…ғзҙ 
print(names[1::2])
# иҙҹж•°жӯҘй•ҝиЎЁзӨәд»ҺеҗҺеҲ°еүҚ
print(names[::-1])
```

```python
# 库存数据格式：（货物名称, 当前库存, 安全阈值）  
inventory_data = [  
    ("包装箱", 850, 500),  
    ("托盘", 180, 200),  
    ("安全帽", 3000, 1000)  
]  
inventory_data[1] = ("托盘", 150, 200)
# 检查并预警低库存  
for item in inventory_data:  
    name, stock, threshold = item  
    if stock < threshold:  
        print(f"警告！{name}库存不足（当前：{stock}/最低：{threshold}）")  
```

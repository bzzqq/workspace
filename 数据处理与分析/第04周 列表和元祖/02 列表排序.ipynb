```python
# 案例1：按配送距离排序（单位：公里）
distances = [15.3, 8.2, 5.7, 12.1, 3.9]
distances.sort()
print("最近配送顺序：", distances)  # 输出：[3.9, 5.7, 8.2, 12.1, 15.3]
```

```python
# 案例2：由远到近排序（单位：公里）
distances = [15.3, 8.2, 5.7, 12.1, 3.9]
distances.sort(reverse=True)
print("由远到近排序：", distances)  # 输出：[15.3, 12.1, 8.2, 5.7, 3.9]
```

```python
# 案例2：按订单号紧急程度排序（1-普通，2-加急）
urgency_levels = [1, 2, 1, 2, 1]
urgency_levels.sort(reverse=True)  # 加急订单优先
print("紧急订单优先：", urgency_levels)  # 输出：[2, 2, 1, 1, 1]
```

```python
# 案例3：按包裹重量排序（单位：kg）
package_weights = [2.5, 5.0, 1.8, 3.2, 4.7]
sorted_weights = sorted(package_weights)
print("轻量优先：", sorted_weights)  # 输出：[1.8, 2.5, 3.2, 4.7, 5.0]
```

```python
# 案例4：按到货时间排序（格式：HHMM）
arrival_times = ['0830', '1545', '0920', '1400', '1030']
arrival_times.sort()
print("到货时间顺序：", arrival_times)  # 输出：['0830', '0920', '1030', '1400', '1545']
```

```python
# 案例5：按仓库温度告警等级排序（3-紧急，2-警告，1-正常）
alarm_levels = [2, 1, 3, 1, 2]
alarm_levels.sort(reverse=True)
print("紧急告警优先：", alarm_levels)  # 输出：[3, 2, 2, 1, 1]
```

```python
# 统计描述
import numpy as np

a = np.array([2, 12, 11, 3, 4, 5, 6, 7, 8, 9, 10])
print(np.sort(a))
print(a.max())  # 最大值
print(a.min())  # 最小值
print(a.sum())  # 求和
print(a.mean())  # 平均值
print(a.std())  # 方差
print(a.var())  # 标准差
```

```python
from numpy import *

L = [[2.73351472, 0.47539713, 3.63280356, 1.4787706, 3.13661701], [1.40305914, 2.27134829, 2.73437132, 1.88939679, 0.0384238], [1.56666697, -0.40088431, 0.54893762, 3.3776724, 2.27490386]]

#将列表转化为数组
arrL = array(L)

## 获取arr1,...arr5
arr1 = arrL[:, 1:2]

arr2 = arrL[1:, [2, 3, 4]]
arr22 = arrL[1:, 2:]

arr3 = arrL[[0, 0, 2, 2], [1, 3, 1, 3]]
arr33 = arrL[::2, [1, 3]]
arr4 = arrL[(arrL >= 2.5) & (arrL <= 3.5)]  ##(arrL[where((arrL>=2.5) & (arrL<=3.5))])
arr5 = arrL[where((arrL < 0) | (arrL > 3))]

print("arr1=", arr1)
print("arr2=", arr2)
print("arr22=", arr22)
print("arr3=", arr3)
print("arr33=", arr33)
print("arr4=", arr4)
print("arr5=", arr5)
```

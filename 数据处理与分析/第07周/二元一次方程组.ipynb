```python
import numpy as np

A = np.array([[3, 2], [1, -1]])  # 系数矩阵
b = np.array([7, -1])  # 常数项
x = np.linalg.solve(A, b)  # 求解
print(x)  # 输出：[1. 2.]
```

```python
from sympy import symbols, solve


def solve_linear_system(a1, b1, c1, a2, b2, c2):
    x, y = symbols('x y')
    eq1 = a1 * x + b1 * y - c1
    eq2 = a2 * x + b2 * y - c2
    solution = solve((eq1, eq2), (x, y))
    return solution


solve_linear_system(1, 1, 0.03, 1 / 4, 3 / 4, 0.0125)
```

```python
import numpy as np


def solve_linear_system(a1, b1, c1, a2, b2, c2):
    A = np.array([[a1, b1], [a2, b2]])
    B = np.array([c1, c2])
    x, y = np.linalg.solve(A, B)
    return {'x': round(x, 2), 'y': round(y, 2)}


solve_linear_system(1, 1, 0.03, 1 / 4, 3 / 4, 0.0125)
```

```python
import numpy as np


def solve_linear_system(a1, b1, c1, a2, b2, c2):
    A = np.array([[a1, b1], [a2, b2]])
    B = np.array([c1, c2])
    x, y = np.linalg.solve(A, B)
    print(f'x={x:.2f}, y={y:.2f}')


solve_linear_system(1, 3, 0.15, 4, 6, 0.4)
```

```python
# 求导 f(x) = 1/x
from sympy import symbols, diff  # 导入符号变量和求导函数

x = symbols('x')  # 定义符号变量x
expr = 1 / x  # 定义表达式1/x
derivative = diff(expr, x)  # 对x求导
print(derivative)  # 输出导数结果
```

```python
# 求导 f(x) = (1-x)/(1+x)
from sympy import symbols, diff  # 导入符号变量和求导函数

x = symbols('x')  # 定义符号变量x
expr = (1 - x) / (1 + x)  # 定义表达式
derivative = diff(expr, x)  # 对x求导
print(derivative)  # 输出导数结果
```

```python
# 求导 f(x) = (1-x)/(1+x)
from sympy import symbols, diff, simplify

x = symbols('x')  # 定义符号变量
f = (1 - x) / (1 + x)  # 定义函数
f_prime = diff(f, x)  # 计算导数
f_prime_simplified = simplify(f_prime)  # 化简
print("导数表达式（化简后）:", f_prime_simplified)
```

```python
# 1/x^2的不定积分
from sympy import symbols, integrate, simplify

x = symbols('x')  # 定义符号变量
integrand = 1 / x**2  # 定义被积函数
result = integrate(integrand, x)  # 计算不定积分
print(result)  # 输出结果
```

```python
# 求3^x*e^xdx的不定积分
from sympy import symbols, integrate, exp, simplify

x = symbols('x') # 定义符号变量
integrand = 3**x * exp(x)  # 定义被积函数
result = integrate(integrand, x) # 计算不定积分
simplified_result = simplify(result)  # 进一步化简
print("积分结果（未化简）:", result) # 输出结果
print("积分结果（化简后）:", simplified_result)
```

```python
# 求x^2dx在区间[0,1]的定积分
from sympy import symbols, integrate

x = symbols('x') # 定义符号变量
result = integrate(x**2, (x, 0, 1)) # 计算定积分
print("定积分结果:", result)  # 输出: 1/3
```

```python
# 一阶线性微分方程y' + y = x求解
from sympy import symbols, Function, dsolve, Eq

x = symbols('x')
y = Function('y')(x)
diff_eq = Eq(y.diff(x) + y, x)  # 方程：y' + y = x
solution = dsolve(diff_eq, y)
print(solution)
```

```python
# 二阶线性微分方程
from sympy import symbols, Function, dsolve, Eq

x = symbols('x')
y = Function('y')(x)
diff_eq = Eq(y.diff(x, 2) + 2*y.diff(x) + 2*y, 0)  # y'' + 2y' + 2y = 0
solution = dsolve(diff_eq, y)
print(solution)
```

```python
# 带初始条件的微分方程
from sympy import symbols, Function, dsolve, Eq

t = symbols('t')
x = Function('x')(t)
diff_eq = Eq(x.diff(t), 1 - x)  # 方程：dx/dt = 1 - x
initial_condition = {x.subs(t, 0): 0}  # 初始条件：x(0)=0
solution = dsolve(diff_eq, x, ics=initial_condition)
print(solution)
```

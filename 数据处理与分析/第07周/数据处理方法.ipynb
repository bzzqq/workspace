```python
# 返回满足条件的数
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
result = arr[arr > 3]  
print(result) # 输出 [4 5 6]
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
result = arr[(arr % 2 == 0) | (arr % 3 == 0)]
print(result) # 输出 [2 3 4 6]
```

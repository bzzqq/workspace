```python
# 计算包含符号的行列式
from sympy import symbols, Matrix

# 定义符号变量
a, b = symbols('a b')
matrix = Matrix([[a, b, b, b], [a, b, a, b], [a, a, b, a], [b, b, b, a]])

# 计算行列式
det = matrix.det()
print("行列式表达式:", det)  
```

```python
# 计算行列式
from sympy import symbols, Matrix

matrix = Matrix([[2, 1, 2], [-4, 3, 1], [2, 3, 5]])

# 计算行列式
det = matrix.det()
print("行列式表达式:", det)
```

```python
# 矩阵乘法
import numpy as np

# 定义矩阵 A 和 B
A = np.array([[1, 2, 3]])
B = np.array([[2, 1], [0, 1], [1, 1]])

# 计算 AB
AB = np.dot(A, B)
print(AB)
```

```python
# 矩阵转置 计算 (AB)^T
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A.T)
```

```python
# 计算矩阵的秩
import numpy as np

# 创建矩阵
A = np.array([[1, 1, 2, 3], [1, 2, 4, 5], [0, 1, 2, 2]])

# 计算秩
rank = np.linalg.matrix_rank(A)
print("矩阵的秩为：", rank)
```

```python
# 计算矩阵的秩
from sympy import Matrix

# 创建矩阵
A = Matrix([[1, 1, 2, 3], [1, 2, 4, 5], [0, 1, 2, 2]])
# 计算秩
rank = A.rank()
print("矩阵的秩为：", rank)  # 输出：2
```

```python
# 计算逆矩阵
import numpy as np

A = np.array([[1, 2, 3], [1, 4, 7], [2, 5, 9]])
A_inv = np.linalg.inv(A)
print("矩阵A的逆矩阵为：", A_inv)
```

```python
# 计算逆矩阵
from sympy import symbols, Matrix

a, b, c, d = symbols('a b c d')
A = Matrix([[a, b], [c, d]])
A_inv = A.inv()
print("矩阵A的逆矩阵为：", A_inv)
```

```python
data = np.random.randn(10, 3)  # 100行3列的随机数据
normalized = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
print(normalized)
```

```python
base_stock = np.array([50, 200, 80])  # 基准库存
adjustment = np.array([[0], [10], [-5]])  # 各仓调整量

# 通过广播实现批量调整
new_stock = base_stock + adjustment
print("调整后库存:\n", new_stock)
```

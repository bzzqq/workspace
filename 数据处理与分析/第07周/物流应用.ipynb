```python
import numpy as np

# 距离矩阵（城市A-B-C）
distance_matrix = np.array([[0, 200, 300],
                            [200, 0, 150],
                            [300, 150, 0]])

# 运输成本矩阵（元/公里）
cost_per_km = np.array([[1.5], [2.0], [1.8]])  # 不同车型成本

# 计算总成本矩阵
total_cost = distance_matrix @ cost_per_km
print("各路径总成本（元）：\n", total_cost)

```

```python
# 需求预测与库存管理
# 基于历史销售数据预测未来需求，优化库存水平。
import numpy as np

# 模拟历史销量数据（12个月）
sales = np.array([120, 135, 145, 160, 180, 200, 210, 190, 170, 150, 130, 110])

# 计算3个月移动平均预测
window_size = 3
moving_avg = np.convolve(sales, np.ones(window_size)/window_size,  mode='valid')
print("移动平均预测：", moving_avg)
```

```python
# 运输成本优化
import numpy as np
from scipy.optimize import linprog

# 成本系数矩阵（车型成本 x 路线成本）
cost_matrix = np.array([[5, 8], [6, 7], [7, 6]])

# 目标函数：最小化总成本（展平矩阵）
c = cost_matrix.flatten()

# 约束条件（如总运输量 >= 需求）
A_ub = -np.eye(6)  # 示例约束
b_ub = -np.array([3, 2, 4, 2, 3, 1])  # 最低运输量

result = linprog(c, A_ub=A_ub, b_ub=b_ub)
print("最优成本分配：", result.x.reshape(3, 2))
```

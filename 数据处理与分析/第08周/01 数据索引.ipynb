```python
# 选择行与列

df = pd.DataFrame(np.random.rand(12).reshape(3, 4) * 100, index=['one', 'two', 'three'], columns=['a', 'b', 'c', 'd'])
print(df)

data1 = df[['a']]
data2 = df[['a', 'c']]
print(data1, type(data1))
print(data2, type(data2))
print('-----')
# 按照列名选择列，只选择一列输出Series，选择多列输出Dataframe

data3 = df.loc[['one']]
data4 = df.loc[['one', 'two']]
print(data3, type(data3))
print(data4, type(data4))
# 按照index选择行，只选择一行输出Series，选择多行输出Dataframe
```

```python
data = {'name': ['Jack', 'Mary', 'Tom'], 'age': [18, 22, 26], 'gender': ['male', 'female', 'male']}
df = pd.DataFrame(data, index=['a', 'b', 'c'])
df
```

```python
data1 = df[['name']]
print(data1, type(data1))
data1
```

```python
data2 = df[['name', 'age']]
print(data2, type(data2))
data2
```

```python
data3 = df[1:]
data3
```

```python
data4 = df.loc[0:1]
data4
```

```python
data5 = df.iloc[0:2]
data5
```

```python
data6 = df.iloc[[1, 0, 2]]
data6
```

```python
data7 = df.iloc[0:2, 0:2]
data7
```

```python
data8 = df.iloc[[0,2], [0,2]]
data8
```

```python
df1 = df[df['age'] > 20]
df1
```

```python
df2 = df[df['gender'] == 'male']
df2
```

```python
df3 = df[(df['gender'] == 'male') & (df['age'] > 20)]
df3
```

```python
df1 = df[['name', 'gender']].iloc[[0, 2]]
df1
```

```python
df1 = df[['name', 'gender']].loc[['a', 'c']]
df1
```

```python
import pandas as pd

# 创建示例问卷数据（假设数据存储在CSV文件中）
data = {"性别": ["男", "女", "女", "男", "女", "男", "女"], "年龄组": ["18-25", "26-35", "18-25", "36-45", "26-35", "18-25", "36-45"], "满意度": ["满意", "非常满意", "一般", "不满意", "满意", "满意", "一般"]}
df = pd.DataFrame(data)

# 统计单选项（如性别、满意度）
print("性别分布：")
print(df["性别"].value_counts())

print("\n满意度分布：")
print(df["满意度"].value_counts(normalize=True))
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'A': ['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3']})
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']})
df3 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'], 'key2': ['K0', 'K1', 'K0', 'K1'], 'A': ['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3']})
df4 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'], 'key2': ['K0', 'K0', 'K0', 'K0'], 'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']})
print(pd.merge(df1, df2, on='key'))
print('------')
# left：第一个df
# right：第二个df
# on：参考键

print(pd.merge(df3, df4, on=['key1', 'key2']))
df4
```

```python
df = pd.DataFrame(np.random.randn(5, 3), index=['a', 'c', 'e', 'f', 'h'], columns=['one', 'two', 'three'])

df = df.reindex(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])

df
```

```python
print(df.isnull())
```

```python
df.dropna()
```

```python
df.fillna(0)  # 全局填充为0
df.fillna(df.mean())  # 按列均值填充
df.fillna(method='ffill')  # 前一个有效值填充
df.fillna(method='bfill')  # 后一个有效值填充
```

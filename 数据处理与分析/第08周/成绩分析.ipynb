```python
import pandas as pd
import numpy as np

df = pd.read_excel("scores.xlsx")
df
```

```python
# 1班语文平均成绩
print(df[df["班级"] == 1]["语文"].mean())
# 1班语文平均成绩，不包含0分
print(df[(df["班级"] == 1) & (df["语文"] != 0)]["语文"].mean())
# 计算总成绩，增加到最后一列
df["总成绩"] = df["语文"] + df["数学"] + df["英语"] + df["物理"] + df["化学"] + df["生物"]
```

```python
# 数据预处理
# 将各科成绩中的0替换为NaN（表示无效值），确保计算平均值时自动忽略这些数据：
score_columns = ['语文', '数学', '英语', '物理', '化学', '生物']
df[score_columns] = df[score_columns].replace(0, np.nan)
df
```

```python
# 生成透视表
# 使用pivot_table按班级分组，计算各科平均分：
# pivot_table = df.pivot_table(index='班级', values=score_columns, aggfunc='mean').round(2)  # 保留两位小数
# 将pivot_table保存到Excel文件：
# pivot_table.to_excel('pivot_table.xlsx')
# pivot_table
score_columns = ['语文', '数学', '英语', '物理', '化学', '生物']
pd.pivot_table(df, index='班级', values=score_columns, aggfunc='mean').round(2)
```

```python
import pandas as pd

# 示例数据
data = {'地区': ['北京', '上海', '北京', '上海', '北京'], '产品类别': ['A', 'B', 'A', 'B', 'A'], '销售额': [200, 150, 300, 250, 400], '利润': [50, 30, 70, 40, 90]}
df = pd.DataFrame(data)

# 创建透视表：按地区和产品类别分组，计算销售额总和与利润均值，填充缺失值为0
pivot_table = pd.pivot_table(df, values=['销售额', '利润'], index='地区', columns='产品类别',
 aggfunc={'销售额': 'sum', '利润': 'mean'}, fill_value=0, margins=True)
pivot_table
```

```python
# 导入seaborn
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('scores.xlsx')
math_scores = df[(df['班级'] == 1) & (df['数学'] != 0)][['数学']]
# plt.hist(math_scores, bins=6, range=(0, 120))
sns.displot(data=math_scores, x="数学", kde=True)
plt.title('数学成绩分布直方图', fontsize=14)
plt.ylabel('人数', fontsize=14)
plt.show()
```

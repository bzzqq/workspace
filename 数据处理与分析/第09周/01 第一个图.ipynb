```python
x = np.linspace(0, np.pi * 2, 100)  # 指定x坐标
y = np.sin(x)  # 指定y坐标
y2 = np.cos(x)  # 指定y2坐标
plt.plot(x, y, 'b-', label='sin(x)')
plt.plot(x, y2, 'r--', label='cos(x)')
plt.legend()
# plt.plot(x, y, x, y2) # 画出sin和cos曲线
# plt.xticks([0, np.pi / 2, np.pi, np.pi * 3 / 2, np.pi * 2], [0, r'$\frac{1}{2}\pi$', r'$\pi$', r'$\frac{3}{2}\pi$', r'$2\pi$'])  # 指定x坐标轴刻度
plt.xticks([0, np.pi / 2, np.pi, np.pi * 3 / 2, np.pi * 2], [0, '1/2π', 'π', '3/2π', '2π'])
plt.yticks([-1.0, -0.5, 0, 0.5, 1.0])
plt.axis([0, np.pi * 2, -1.0, 1.0])
# plt.legend(['sin(x)', 'cos(x)'])
plt.text(3.0, 0.2, 'y = sin(x)')
plt.show()  # 显示图像
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x - 0), color='blue')  # 通过颜色名称指定
plt.plot(x, np.sin(x - 1), color='g')  # 通过颜色简写名称指定(rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')  # 介于0-1之间的灰阶值
plt.plot(x, np.sin(x - 3), color='#FFDD44')  # 16进制的RRGGBB值
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))  # RGB元组的颜色值，每个值介于0-1
plt.plot(x, np.sin(x - 5), color='chartreuse')  # 能支持所有HTML颜色名称值
plt.show()
```

```python
# 一周温度变化的折线图

# 模拟一周温度变化的数据
temperatures = [26, 27, 29, 28, 26, 24, 25]

# 生成时间轴
time = range(1, len(temperatures) + 1)

# 创建折线图
plt.plot(time, temperatures)

# 设置图表标题并添加标签
plt.title("一周温度变化")
plt.show()
```

```python
# 1.模拟数据准备（假设30天数据）
days = np.arange(1, 31)
np.random.seed(70)
order_volume = np.random.randint(50, 200, size=30)  # 随机生成50-200单/日
# 2.创建画布
plt.figure(figsize=(12, 6), dpi=100)
plt.title("每日订单处理趋势（2025年4月）", fontsize=14, pad=20)
plt.xlabel("日期", fontsize=12)
plt.ylabel("订单处理量", fontsize=12)
plt.grid(alpha=0.3)

# 3.绘制折线图
line = plt.plot(days, order_volume, color='#2E86C1', linewidth=2, marker='o', markersize=8, markeredgecolor='white', markerfacecolor='#2E86C1')

# 4.标注极值点
max_day = days[np.argmax(order_volume)]
plt.annotate(f'峰值: {max(order_volume)} orders', xy=(max_day, max(order_volume)), \
    xytext=(max_day + 2, max(order_volume) - 30), arrowprops=dict(arrowstyle='->', color='gray'))

# 5.样式优化
plt.xticks(days)
plt.tight_layout()
plt.show()
```

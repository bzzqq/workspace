```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 散点图
orders = [128, 164, 74, 110, 109, 178, 167, 135, 167, 134, 153, 73, 102, 134,\
     164, 168, 179, 194, 91, 127, 140, 199, 97, 110, 148, 65, 60, 113, 156, 189]
days = np.arange(1, 31)
plt.figure(figsize=(12, 6))  # 宽度12英寸，高度6英寸
plt.plot(orders, 'o')
plt.grid(alpha=0.3)
plt.xticks(days)
plt.tight_layout()
```

```python
import random

random.seed(1)

# 生成男性数据
male_heights = [random.randint(165, 185) for _ in range(20)]
male_weights = [random.randint(60, 80) for _ in range(20)]

# 生成女性数据
female_heights = [random.randint(150, 175) for _ in range(20)]
female_weights = [random.randint(45, 65) for _ in range(20)]

# 绘制散点图
plt.figure(figsize=(10, 6), dpi=300)
# plt.scatter(male_heights, male_weights, color='blue', label='男性')
# plt.scatter(female_heights, female_weights, color='red', label='女性')
plt.plot(male_heights, male_weights, 'o', label='男性')
plt.scatter(female_heights, female_weights, marker='o', color='orange', label='女性')

# 设置图表标题和坐标轴标签
plt.title('身高与体重的散点图')
plt.xlabel('身高 (cm)')
plt.ylabel('体重 (kg)')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T
plt.figure(figsize=(8, 5), dpi=300)
plt.scatter(features[0], features[1], alpha=0.2, s=100 * features[3], c=iris.target, cmap='viridis')
plt.xlabel('花萼长度')
plt.ylabel('花萼宽度')
plt.title('鸢尾花数据')
plt.colorbar()  # 显示颜色对比条
plt.show()
```

```python
df = pd.read_excel('scores.xlsx')
x = np.arange(5)
y1 = df.head()['语文']
y2 = df.head()['数学']
y3 = df.head()['英语']
labels = df.head()['姓名']
plt.figure(figsize=(10, 8), dpi=300)
bar_width = 0.25
bar1 = plt.bar(x, y1, width=bar_width, label='语文')
bar2 = plt.bar(x + bar_width, y2, tick_label=labels, width=bar_width, label='数学')
bar3 = plt.bar(x + 2 * bar_width, y3, width=bar_width, label='英语')
bars = [bar1, bar2, bar3]
ys = [y1, y2, y3]
for i in range(3):
    plt.bar_label(
        bars[i],
        labels=ys[i],
        label_type='edge',  # 标签位置：顶部
        padding=3  # 标签与条形的间距
    )
plt.legend()
plt.title('第二次月考成绩')
plt.show()
```

```python
df = pd.read_excel('scores.xlsx')
x = np.arange(5)
y1 = df.head()['语文']
y2 = df.head()['数学']
y3 = df.head()['英语']
labels = df.head()['姓名']
plt.figure(figsize=(10, 8), dpi=300)
bar1 = plt.bar(x, y1, label='语文', tick_label=labels)
bar2 = plt.bar(x, y2, label='数学', bottom=y1)
bar3 = plt.bar(x, y3, label='英语', bottom=y1 + y2)
bars = [bar1, bar2, bar3]
ys = [y1, y2, y3]
for i in range(3):
    plt.bar_label(
        bars[i],
        labels=ys[i],
        label_type='center',  # 标签位置：中间
        padding=3  # 标签与条形的间距
    )
plt.legend()
plt.title('第二次月考成绩')
plt.show()
```

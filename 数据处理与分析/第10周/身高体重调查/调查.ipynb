```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('身高体重调查.xlsx')
x = df['2、您的身高（厘米）：']
y = df['3、您的体重（千克）：']
gender = df['1、您的性别：']
shoe_size = df['4、您的鞋码：']
plt.scatter(x, y, c=gender, s=shoe_size, alpha=0.5)
plt.show()
df
```

```python
df = pd.read_excel('身高体重调查.xlsx')
df = df[(df['鞋码'] < 50) & (df['身高'] < 190)]
x = df['身高']
y = df['体重']
gender = df['性别']
shoe_size = df['鞋码']
plt.scatter(x, y, c=gender, s=shoe_size, alpha=0.5)
plt.show()
df
```

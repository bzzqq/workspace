```python
def cosmic_scale():
    """展示不同尺度的数字"""
    print("🌌 宇宙尺度展示")
    
    # 极小到极大
    planck_length = 1.6e-35  # 普朗克长度（米）
    proton_size = 0.83e-15    # 质子大小（米）
    human_height = 1.75e0     # 人的身高（米）
    earth_diameter = 1.27e7   # 地球直径（米）
    solar_system = 2.9e14     # 太阳系直径（米）
    observable_universe = 8.8e26  # 可观测宇宙直径（米）
    
    print(f"普朗克长度: {planck_length:.2e} 米")
    print(f"质子大小: {proton_size:.2e} 米")
    print(f"人的身高: {human_height:.2e} 米")
    print(f"地球直径: {earth_diameter:.2e} 米")
    print(f"太阳系直径: {solar_system:.2e} 米")
    print(f"可观测宇宙: {observable_universe:.2e} 米")
    
    # 比较
    ratio = observable_universe / human_height
    print(f"\n人与宇宙的比例: 1 : {ratio:.2e}")

cosmic_scale()
```

```python
# 使用基本运算计算年龄相关数字
birth_year = 2010
current_year = 2024

age = current_year - birth_year
age_in_months = age * 12
age_in_days = age * 365
age_in_hours = age_in_days * 24

print("🎂 生日计算器")
print("-" * 30)
print(f"出生年份: {birth_year}")
print(f"当前年份: {current_year}")
print(f"年龄: {age} 岁")
print(f"大约活了: {age_in_months} 个月")
print(f"大约活了: {age_in_days} 天")
print(f"大约活了: {age_in_hours} 小时")
```

```python
# 购物结算
print("🛒 购物车结算")
print("-" * 30)

# 商品价格
book_price = 45.50
pen_price = 3.50
notebook_price = 12.00

# 购买数量
book_qty = 2
pen_qty = 5
notebook_qty = 3

# 小计
book_subtotal = book_price * book_qty
pen_subtotal = pen_price * pen_qty
notebook_subtotal = notebook_price * notebook_qty

# 总计
subtotal = book_subtotal + pen_subtotal + notebook_subtotal
tax_rate = 0.08  # 8%税率
tax = subtotal * tax_rate
total = subtotal + tax

print("商品明细:")
print(f"书籍 ×{book_qty}: ¥{book_subtotal:.2f}")
print(f"钢笔 ×{pen_qty}: ¥{pen_subtotal:.2f}")
print(f"笔记本 ×{notebook_qty}: ¥{notebook_subtotal:.2f}")
print(f"小计: ¥{subtotal:.2f}")
print(f"税费(8%): ¥{tax:.2f}")
print(f"总计: ¥{total:.2f}")
```

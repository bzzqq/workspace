```python
print(int("5"))  # int()可以将整数字符串转换成整数
print(int(3.14)) # int()可以将浮点数转换成整数
print(int(9.8))  # int()是截断不是四舍五入
print(int("3.14")) # int()不能将小数字符串转换成整数
```

```python
print(float(5))      # float()可以将整数转换成浮点数
print(float("5"))    # float()可以将整数字符串转换成浮点数
print(float("3.14")) # float()可以将浮点数字符串转换成浮点数
```

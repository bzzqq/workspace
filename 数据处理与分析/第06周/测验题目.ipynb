```python
# 计算圆的体积
import math

radius = float(input())

def calculate_circle_area(r):
    area = 4 / 3 * math.pi * radius**3
    return round(area, 2)
print(calculate_circle_area(radius))
```

```python
# 谁没有及格？
students = {"小明": 50, "小红": 85, "小刚": 42, "小李": 48, "小赵": 95, "小钱": 87, "小周": 83}
def get_mvp(students):
    mvp = {}
    for name, score in students.items():
        if score < 60:
            mvp[name] = score
    return mvp
print(get_mvp(students))
```

```python
import pandas as pd
import random
import os

# 加载学生数据
def load_students():
    df = pd.read_excel('点名册.xlsx')
    return df[['序号', '学号', '姓名', '班级']].to_dict('records')

# 主程序
def main():
    # 加载数据
    students = load_students()
    total = len(students)
    
    # 复制一份用于点名
    remaining = students.copy()
    called = []
    
    print("=" * 50)
    print("        课堂点名程序")
    print("=" * 50)
    print(f"总人数：{total}")
    print("-" * 50)
    
    while True:
        print("\n请选择操作：")
        print("1. 随机点名")
        print("2. 查看已点名名单")
        print("3. 查看未点名名单")
        print("4. 重置点名")
        print("5. 退出")
        
        choice = input("\n请输入数字选择：").strip()
        
        if choice == '1':
            if not remaining:
                print("\n⚠️ 所有人都点过名了！请先重置。")
                continue
            
            # 随机选一个
            student = random.choice(remaining)
            
            # 移入已点名列表
            remaining.remove(student)
            called.append(student)
            
            # 显示结果
            print("\n" + "=" * 50)
            print(f"🎉 被点到的同学：{student['姓名']}")
            print(f"   班级：{student['班级']}")
            print(f"   学号：{student['学号']}")
            print("=" * 50)
            print(f"已点名：{len(called)}人 | 剩余：{len(remaining)}人")
            
        elif choice == '2':
            if not called:
                print("\n📭 还没有人点过名")
            else:
                print(f"\n📋 已点名名单 ({len(called)}人)：")
                print("-" * 60)
                for i, s in enumerate(called, 1):
                    print(f"{i:2d}. {s['姓名']} - {s['班级']} ({s['学号']})")
                    
        elif choice == '3':
            if not remaining:
                print("\n✅ 所有人都点过名了！")
            else:
                print(f"\n📋 未点名名单 ({len(remaining)}人)：")
                print("-" * 60)
                for i, s in enumerate(remaining, 1):
                    print(f"{i:2d}. {s['姓名']} - {s['班级']}")
                    
        elif choice == '4':
            remaining = students.copy()
            called = []
            print("\n🔄 点名记录已重置")
            
        elif choice == '5':
            print("\n👋 程序退出")
            break
            
        else:
            print("\n❌ 输入错误，请重新选择")

if __name__ == "__main__":
    try:
        main()
    except FileNotFoundError:
        print("❌ 错误：找不到 '点名册.xlsx' 文件")
        print("请确保文件在当前目录下")
    except Exception as e:
        print(f"❌ 错误：{e}")
```

```python
name = "张三"
age = "18"

print("姓名：", name, "年龄：", age)
# 输出：姓名： 张三 年龄： 18（注意有空格）

print("姓名："+ name+ "年龄："+ age)
# 输出：姓名： 张三 年龄： 18（注意有空格）

# 去掉空格可以用 sep
print("姓名：", name, "年龄：", age, sep="")
# 输出：姓名：张三年龄：18
```

```python
name = "张三"
age = 18
score = 95.5

# 1. 用逗号（print自动加空格）
print("姓名：", name, "，年龄：", age, "，分数：", score)
# 输出：姓名： 张三 ，年龄： 18 ，分数： 95.5

# 去掉空格（用sep参数）
print("姓名：", name, "，年龄：", age, "，分数：", score, sep="")
# 输出：姓名：张三，年龄：18，分数：95.5

# 2. 用加号（需要转换类型）
print("姓名：" + name + "，年龄：" + str(age) + "，分数：" + str(score))
# 输出：姓名：张三，年龄：18，分数：95.5

# 3. 用f-string（最简洁）
print(f"姓名：{name}，年龄：{age}，分数：{score}")
# 输出：姓名：张三，年龄：18，分数：95.5
```

```python
name = "张三"
age = 18
score = 95.5

# 1. 用逗号（用sep=""去掉空格）
print("姓名：", name, "，年龄：", age, "，分数：", score, sep="")
# 2. 用加号（需要转换类型）
print("姓名：" + name + "，年龄：" + str(age) + "，分数：" + str(score))
# 3. 用f-string（最简洁）
print(f"姓名：{name}，年龄：{age}，分数：{score}")
```

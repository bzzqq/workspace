```python
text = "zhang Bin"

print(text.upper())       # 输出: ZHANG BIN（全大写）
print(text.lower())       # 输出: zhang bin（全小写）
print(text.capitalize())  # 输出: Zhang bin（首字母大写）
print(text.title())       # 输出: Zhang Bin（每个单词首字母大写）
```

```python
text = "   hello   "
print(text.strip())   # 输出: "hello"（去掉首尾空格）
print(text.lstrip())  # 输出: "hello   "（去掉左边空格）
print(text.rstrip())  # 输出: "   hello"（去掉右边空格）
```

```python
# 常见的空白字符
' '      # 空格
'\t'     # 制表符（Tab）
'\n'     # 换行符
'\r'     # 回车符
'\f'     # 换页符
'\v'     # 垂直制表符

# 示例：这些看起来都是"空白"
text = "Hello\tWorld\nPython"
print(text)  # Hello	World
             # Python
```

```python
# strip() 可以指定要去除的字符
text = "###Hello World###"
print(text.strip('#'))        # 输出: 'Hello World'

# 去除多种字符
text2 = "***Hello World!!!"
print(text2.strip('*!'))      # 输出: 'Hello World'

text3 = "abcHello Worldcba"
print(text3.strip('abcde'))     # 输出: 'Hello Worl'
# 解释：去掉了首尾所有的 a,b,c,d,e，直到遇到不是a/b/c/d/e的字符
```

```python
text = "I like Python, Python is fun"

# 查找位置
print(text.find("Python"))   # 输出: 7（"Python"第一次出现的位置）
print(text.find("Java"))     # 输出: -1（找不到返回-1）

# 替换
print(text.replace("Python", "Java"))  
# 输出: "I like Java, Java is fun"

# 数一数有多少个
print(text.count("Python"))  # 输出: 2
```

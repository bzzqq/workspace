```python
text = "Python"
print(text[0])  # 输出: P（第1个字符，编号0）
print(text[1])  # 输出: y
print(text[5])  # 输出: n（第6个字符，编号5）
```

```python
text = "HelloWorld"
print(text[0:5])  # 输出: Hello（位置0到4，不包括5）
print(text[5:10])  # 输出: World（位置5到9）
print(text[:5])  # 输出: Hello（开头到位置5）
print(text[5:])  # 输出: World（位置5到末尾）
print(text[:])  # 输出: HelloWorld（整个字符串）
```

```python
sentence = "我喜欢吃苹果"
print("苹果" in sentence)  # 输出: True（存在）
print("香蕉" in sentence)  # 输出: False（不存在）
print("西瓜" not in sentence)  # 输出: True（确实不存在）
```

```python
mood = ["开心", "烦恼", "开心", "烦恼", "开心", "烦恼", "开心"]

while "烦恼" in mood:
    mood.remove("烦恼")
print(mood)
```

```python
mood = ["开心", "烦恼", "开心", "烦恼", "开心", "烦恼", "开心"]

new_mood = []
for item in mood:
    if item != "烦恼":
        new_mood.append(item)
print(new_mood)
```

```python
ring_lst = list(range(1, 15))# 构造一个元素为1到14的列表
while len(ring_lst) >= 3:  # 列表长度大于等于3时
    ring_lst = ring_lst[3:] + ring_lst[:2]
    print(ring_lst)
    # 前2个元素拼接到第3个元素后的列表末尾
print(ring_lst)  # 循环外输出
```

```python
lst = list(range(1, 15))# 构造一个元素为1到14的列表
while len(lst) >= 3:  # 列表长度大于等于3时
    # 前2个元素拼接到第3个元素后的列表末尾
    lst = lst[3:] + lst[:2]
print(lst)  # 循环外输出
```

```python
lst = list(range(1, 15))# 构造一个元素为1到14的列表
while len(lst) >= 3:  # 列表长度大于等于3时
    # 前2个元素拼接到第3个元素后的列表末尾
    lst = lst[3:] + lst[:2]
print(sorted(lst))  # 循环外输出
```

```python
print(int("123"))    # 输出: 123
print(float("123"))  # 输出: 123.0
```

```python
# 字符串 → 列表（每个字符变成元素）
print(list("Python"))    # 输出: ['P', 'y', 't', 'h', 'o', 'n']

# 元组 → 列表
print(list((1, 2, 3)))   # 输出: [1, 2, 3]

# 范围 → 列表
print(list(range(5)))    # 输出: [0, 1, 2, 3, 4]
```

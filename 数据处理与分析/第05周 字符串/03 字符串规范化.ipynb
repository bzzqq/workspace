```python
# 创建字母表
lower = 'abcdefghijklmnopqrstuvwxyz'
upper = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# 创建转换表
shifted_lower = lower[3:] + lower[:3] # 'defghijklmnopqrstuvwxyzabc'
shifted_upper = upper[3:] + upper[:3] # 'DEFGHIJKLMNOPQRSTUVWXYZABC'

text = "Hello World"
result = ""
for char in text:
    if char in lower:
        index = lower.find(char)
        result += shifted_lower[index]
    elif char in upper:
        index = upper.find(char)
        result += shifted_upper[index]
    else:
        result += char
print(result)
```

```python
# 创建字母表
lower = 'abcdefghijklmnopqrstuvwxyz'
upper = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

# 创建转换表
shifted_lower = lower[_:_] + lower[_:_]  # 'xyzabcdefghijklmnopqrstuvw'
shifted_upper = upper[_:_] + upper[_:_]  # 'XYZABCDEFGHIJKLMNOPQRSTUVW'
text = "Khoor Zruog"
result = ""
for char in text:
    if char in lower:
        index = lower.find(char)
        result += shifted_lower[index]
    elif char in upper:
        index = upper.find(char)
        result += shifted_upper[index]
    else:
        result += char
print(result)
```

```python
text = "I like Python, Python is fun"

# 查找位置
print(text.find("Python"))   # 输出: 7（"Python"第一次出现的位置）
print(text.find("Java"))     # 输出: -1（找不到返回-1）

# 替换
print(text.replace("Python", "Java"))  
# 输出: "I like Java, Java is fun"

# 数一数有多少个
print(text.count("Python"))  # 输出: 2
```

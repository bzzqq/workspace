```python
import pandas as pd

df = pd.read_excel('2025化工制药系毕业答辩成绩.xlsx')
# 按照专业分组，按平均分汇总各区间的人数，优秀>=90，良好>=80，及格<80

# 按专业统计总人数（直接统计专业下的所有记录数）
total_students = df.groupby('专业').size().reset_index(name='总人数')

# 按专业分组统计各等级人数
grouped = df.groupby('专业')['平均分'].agg(
    优秀人数=lambda x: (x >= 90).sum(),
    优良人数=lambda x: (x >= 80).sum(),
    及格人数=lambda x: (x < 80).sum()
).reset_index()

# 合并数据并计算占比（保留两位小数）
merged = pd.merge(grouped, total_students, on='专业')
merged['优秀占比'] = (merged['优秀人数'] / merged['总人数']).map('{:.2%}'.format)
merged['优良占比'] = (merged['优良人数'] / merged['总人数']).map('{:.2%}'.format)
merged['及格占比'] = (merged['及格人数'] / merged['总人数']).map('{:.2%}'.format)

# 调整列顺序并输出结果
result = merged[['专业', '优秀人数', '优秀占比', '优良人数', '优良占比', '及格人数', '及格占比', '总人数']]
result
```

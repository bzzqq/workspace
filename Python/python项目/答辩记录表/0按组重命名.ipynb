```python
import pandas as pd
import os


df = pd.read_excel('2025化工制药系毕业答辩分组.xlsx')
# 读取当前目录下所有word文件，如果文件名中包含df['学生']列中的数据，则按照df中的信息将文件重命名为答辩组-指导教师-学号-学生。
for file in os.listdir():
    if file.endswith('.docx') or file.endswith('.doc'):
        for name in df['学生']:
            if name in file:
                # 获取匹配的行
                matched_row = df[df['学生'] == name].iloc[0]
                # 构造新文件名
                new_name = f"{matched_row['答辩组']}-{matched_row['学号']}-{matched_row['学生']}{os.path.splitext(file)[1]}"
                os.rename(file, new_name)
                print(file, '重命名为', new_name)
```

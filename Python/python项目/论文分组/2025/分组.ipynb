```python
import os
import shutil


def rename_and_move_files(file_mapping, base_path):
    for group, students in file_mapping.items():
        for index, student in enumerate(students, start=1):
            student_id = student['学号']
            student_name = student['学生']
            title = student['题目']
            new_name = f"{group}组{index}号_{student_name}_{title}"

            # 处理论文初稿文件
            draft_file = [f for f in os.listdir(os.path.join(base_path, '初稿')) if student_id in f]
            if draft_file:
                draft_file = draft_file[0]
                new_draft_path = os.path.join(base_path, '论文初稿', group, new_name + os.path.splitext(draft_file)[1])
                os.makedirs(os.path.dirname(new_draft_path), exist_ok=True)
                shutil.move(os.path.join(base_path, '初稿', draft_file), new_draft_path)

            # 处理附件文件
            attachment_file = [f for f in os.listdir(os.path.join(base_path, '附件')) if student_id in f]
            if attachment_file:
                attachment_file = attachment_file[0]
                new_attachment_path = os.path.join(base_path, '附件', group, new_name + os.path.splitext(attachment_file)[1])
                os.makedirs(os.path.dirname(new_attachment_path), exist_ok=True)
                shutil.move(os.path.join(base_path, '附件', attachment_file), new_attachment_path)


def main():
    base_path = '.'
    # 构建学生信息映射
    file_mapping = {}
    import openpyxl
    wb = openpyxl.load_workbook('2025化工制药系毕业答辩分组.xlsx')
    sheet = wb['排序分组']
    for row in sheet.iter_rows(min_row=2, values_only=True):
        group = row[0]
        student_id = row[2]
        student_name = row[1]
        title = row[5]
        if group not in file_mapping:
            file_mapping[group] = []
        file_mapping[group].append({'学号': student_id, '学生': student_name, '题目': title})

    rename_and_move_files(file_mapping, base_path)


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']
# 用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
# 读取文件
excel_file = pd.ExcelFile('2018.xlsx')

# 获取指定工作表中的数据
df = excel_file.parse('2020')

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 定义分数段
bins = list(range(63, 101, 3))

# 绘制直方图
plt.figure(figsize=(10, 6))
plt.hist(df.iloc[:, 0], bins=bins, edgecolor='black')
plt.title('过程考核成绩分布直方图')
plt.xlabel('成绩区间')
plt.xticks(rotation=45)
plt.ylabel('数量')
plt.xticks(bins)

plt.show()
```

### 有两列成绩的

```python
import pandas as pd
import re

# 预处理文件：确保姓名和性别之间有空格分隔
with open('基础学科综训(制药).txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

processed_lines = []
for line in lines:
    # 使用正则表达式在姓名和性别之间插入空格
    # 匹配模式：任意字符（非贪婪）后跟"男"或"女"，且其后是数字（班级）
    processed_line = re.sub(r'([^男女生]{2,})(男|女)(\d+)', r'\1 \2 \3', line)
    processed_lines.append(processed_line)

# 将处理后的内容写入临时文件
temp_file = 'temp_student_data.txt'
with open(temp_file, 'w', encoding='utf-8') as f:
    f.writelines(processed_lines)

# 使用pandas读取预处理后的文件
df = pd.read_csv(
    temp_file,
    sep=r'\s+|(?<=\d)(?=[^\d\s])|(?<=[^\d\s])(?=\d)',
    header=None,
    engine='python',
    names=['序号', '学号', '姓名', '性别', '班级', '专业', '平时成绩', '总成绩'],
    dtype={'学号': 'object'}  # 关键修改：指定学号为文本类型
)

# 保存为Excel文件
df.to_excel('基础学科综训(制药).xlsx', index=False)
print("数据已保存")
```
